```python

r = ['Percentage Contribution to Total GDP']
raw_data = {'Agriculture, forestry and fishing': [40499397], 'Real estate, ownership of dwelling & professio': [21517411],
            'Manufacturing': [20618199],'Trade & repair services' : [20527614], 'Construction': [18372797], 'Transport, storage, communication & services': [13325998],
            'Other services': [13111210],'Public administration': [10989309], 'Financial services': [5695704],'Electricity, gas, water supply & other utility.': [3037359]}
df = pd.DataFrame(raw_data)

totals = [i+j+k+l+m+n+o+p+q+r for i,j,k,l,m,n,o,p,q,r in zip(df['Agriculture, forestry and fishing'], df['Real estate, ownership of dwelling & professio'],
            df['Manufacturing'],df['Trade & repair services'], df['Construction'], df['Transport, storage, communication & services'],df['Other services'],df['Public administration'],df['Financial services'],df['Electricity, gas, water supply & other utility.'])]

Agriculture_Forestry_Fishing = [i / j * 100 for i,j in zip(df['Agriculture, forestry and fishing'], totals)]
Real_Estate = [i / j * 100 for i,j in zip(df['Real estate, ownership of dwelling & professio'], totals)]
Manufacturing = [i / j * 100 for i,j in zip(df['Manufacturing'], totals)]
Trade_repair_services = [i / j * 100 for i,j in zip(df['Trade & repair services'], totals)]
Construction = [i / j * 100 for i,j in zip(df['Construction'], totals)]
Transport_storage_communication = [i / j * 100 for i,j in zip(df['Transport, storage, communication & services'], totals)]
Other = [i / j * 100 for i,j in zip(df['Other services'], totals)]
Public_administration = [i / j * 100 for i,j in zip(df['Public administration'], totals)]
Financial_services = [i / j * 100 for i,j in zip(df['Financial services'], totals)]
Electricity_gas_water_supply = [i / j * 100 for i,j in zip(df['Electricity, gas, water supply & other utility.'], totals)]


plt.bar(r, Agriculture_Forestry_Fishing, edgecolor='white', width=0.5)
plt.bar(r, Real_Estate, edgecolor='white', width=0.5)
plt.bar(r, Manufacturing, edgecolor='white', width=0.5)
plt.bar(r, Trade_repair_services, edgecolor='white', width=0.5)
plt.bar(r, Construction, edgecolor='white', width=0.5)
plt.bar(r, Transport_storage_communication, edgecolor='white', width=0.5)
plt.bar(r, Other, edgecolor='white', width=0.5)
plt.bar(r, Public_administration, edgecolor='white', width=0.5)
plt.bar(r, Financial_services, edgecolor='white', width=0.5)


plt.xticks(r)
plt.legend(['Agriculture_Forestry_Fishing','Real_Estate','Manufacturing','Trade_repair_services','Construction','Transport_storage_communication',
          'Other','Public_administration','Financial_services'], fontsize=7.5, loc='best')


plt.show()
```
